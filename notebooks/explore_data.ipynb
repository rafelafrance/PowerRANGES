```python
from pathlib import Path
from zipfile import ZipFile

import pandas as pd
```

## Constants

```python
DATA = Path("..") / "data"

FIELDS = """
    dynamicProperties fieldNotes occurrenceRemarks reproductiveCondition
    """.split()
```

## Examine data contents

```python
zips = sorted(DATA.glob("*.zip"))

for zip in zips:
    print(zip.stem)
```

```python
with ZipFile(zips[0]) as zippy:
    for name in zippy.infolist():
        print(name)
```

```python
with ZipFile(zips[0]) as zippy:
    name = zippy.infolist()[0]
    with zippy.open(name) as zin:
        df = pd.read_csv(zin, sep="\t", dtype="string")
        cols = sorted(df.columns)

cols
```

## Look at contents of target fields

```python
with ZipFile(zips[0]) as zippy:
    name = zippy.infolist()[0]
    with zippy.open(name) as zin:
        df = pd.read_csv(zin, usecols=FIELDS, sep="\t", dtype="string").fillna("")

df.head()
```
